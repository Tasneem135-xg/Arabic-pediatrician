```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import nltk
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_excel (r'cleaning_1.xlsx')
df.shape
```

```python
dfx=df[0:3]
```

```python
dfx
```

```python
dfx['tokenized'] = dfx.apply(lambda row: nltk.word_tokenize(row['question']), axis=1)
```

```python
dfx['tokenized'][0]
```

```python
dfx.shape[0]
```

```python
tokens=[]
question=[]
qid=[]
intents=[]
for rows in range((dfx.shape)[0]):
    #print(dfx['tokenized'][rows])
    
    for i in dfx['tokenized'][rows]:
        #print(i)
        tokens.append(i)
        #print(dfx['question'][rows])
        q=dfx['question'][rows]
        ids=dfx['new_id'][rows]
        intent=dfx['intent'][rows]
        question.append(q)
        qid.append(ids)
        intents.append(intent)
```

```python
data = {'question':question, 'tokens': tokens,'Q_ID':qid,'intent':intents}  
  
# Create DataFrame  
dfc = pd.DataFrame(data) 

dfc.head(30)
```
