```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import nltk
from nltk.tokenize import word_tokenize
```

```python
drug = pd.read_excel (r'D:\df_drug.xlsx')

food = pd.read_excel (r'D:\df_food.xlsx')

symptoms = pd.read_excel (r'D:\df_symptom.xlsx')

test = pd.read_excel (r'D:\df_test.xlsx')

vaccine = pd.read_excel (r'D:\df_vaccine.xlsx')
```

```python
drug
```

```python
intlist=[drug,food,symptoms,test,vaccine]

for i in intlist:
    print(i['new_id'].duplicated().any())
```

```python
drug['tokenized'] = drug.apply(lambda row: nltk.word_tokenize(row['question']), axis=1)

food['tokenized'] = food.apply(lambda row: nltk.word_tokenize(row['question']), axis=1)

symptoms['tokenized'] = symptoms.apply(lambda row: nltk.word_tokenize(row['question']), axis=1)

test['tokenized'] = test.apply(lambda row: nltk.word_tokenize(row['question']), axis=1)

vaccine['tokenized'] = vaccine.apply(lambda row: nltk.word_tokenize(row['question']), axis=1)
```

```python
print(vaccine['tokenized'][0][0])
```

```python
tokens=[]
question=[]
qid=[]
intents=[]

dfx=food
for rows in range((dfx.shape)[0]):
    #print(dfx['tokenized'][rows])
    
    for i in dfx['tokenized'][rows]:
        #print(i)
        tokens.append(i)
        #print(dfx['question'][rows])
        q=dfx['question'][rows]
        ids=dfx['new_id'][rows]
        intent=dfx['intent'][rows]
        question.append(q)
        qid.append(ids)
        intents.append(intent)
        
data = {'question':question, 'tokens': tokens,'Q_ID':qid,'intent':intents}  
  
# Create DataFrame  
dff_food = pd.DataFrame(data) 

dff_food.head()
```
